```python
# Importación de librerías necesarias para análisis estadístico y visualización
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configuración de la primera curva (distribución normal)
m, s = 0, np.sqrt(0.2)  # media m = 0, desviación estándar s = raíz cuadrada de 0.2
x = np.arange(-5, 5, 0.01)  # rango de valores para X
px = norm.pdf(x, m, s)  # cálculo de la función de densidad de probabilidad (pdf)
plt.plot(x, px, linestyle="-", color="blue", linewidth=3)  # dibujando la curva

# Configuración de la segunda curva (distribución normal)
m, s = 2, 1  # media m = 2, desviación estándar s = 1
px = norm.pdf(x, m, s)  # recalculando pdf con nuevos parámetros
plt.plot(x, px, linestyle="-", color="blue", linewidth=3)  # dibujando la curva

# Configuración y dibujo de las curvas tercera a octava
# Similar al proceso anterior, cambiando los valores de media y desviación estándar
# Las curvas tercera a sexta se dibujan en azul y las curvas séptima y octava en rojo

# Tercera curva: media = 0, desviación estándar = raíz cuadrada de 5
m, s = 0, np.sqrt(5)
px = norm.pdf(x, m, s)
plt.plot(x, px, linestyle="-", color="blue", linewidth=3)

# Cuarta curva: media = -2, desviación estándar = raíz cuadrada de 0.5
m, s = -2, np.sqrt(0.5)
px = norm.pdf(x, m, s)
plt.plot(x, px, linestyle="-", color="blue", linewidth=3)

# Quinta curva: media = 1, desviación estándar = raíz cuadrada de 0.2, color rojo
m, s = 1, np.sqrt(0.2)
px = norm.pdf(x, m, s)
plt.plot(x, px, linestyle="-", color="red", linewidth=3)

# Sexta curva: media = 3, desviación estándar = 1, color rojo
m, s = 3, 1
px = norm.pdf(x, m, s)
plt.plot(x, px, linestyle="-", color="red", linewidth=3)

# Séptima curva: media = 1, desviación estándar = raíz cuadrada de 5, color rojo
m, s = 1, np.sqrt(5)
px = norm.pdf(x, m, s)
plt.plot(x, px, linestyle="-", color="red", linewidth=3)

# Octava curva: media = -1, desviación estándar = raíz cuadrada de 0.5, color rojo
m, s = -1, np.sqrt(0.5)
px = norm.pdf(x, m, s)
plt.plot(x, px, linestyle="-", color="red", linewidth=3)

# Configuración del gráfico
plt.title("Estimador de Máxima Verosimilitud Sesgado")  # título del gráfico
plt.legend([r"$s^2=0.2$", r"$s^2=1$", r"$s^2=5$", r"$s^2=0.5$", r"$\sigma^2$"])  # leyenda
plt.xlabel(r"$X$")  # etiqueta del eje X
plt.grid()

# Mostrar el gráfico
plt.show()
```
