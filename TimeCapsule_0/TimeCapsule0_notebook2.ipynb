```python
import math
import numpy as np
import matplotlib.pyplot as plt
from random import random, uniform
```

```python
def inverseGeneration(a, c, b, n=100000):
    """
    Genera una lista de números siguiendo una distribución triangular mediante el método de la inversión.

    Args:
    a, b, c (float): Parámetros de la distribución triangular (a < b < c).
    n (int): Número de valores a generar.

    Returns:
    list: Lista de números generados según la distribución triangular.
    """
    lista = []
    for i in np.random.rand(n):
        num = ((c - a) / (b - a))
        if i < num:
            sqt = (b - a) * (c - a) * i
            x = a + math.sqrt(sqt)
        else:
            sqt = (b - a) * (b - c) * (1 - i)
            x = b - math.sqrt(sqt)
        lista.append(x)
    return lista

# Generación y visualización de un histograma para la distribución triangular inversa
plt.hist(inverseGeneration(-3, 0, 8), bins=200, density=True)
plt.show()
```

```python
def random_triangular(a, b, c):
    """
    Genera un número aleatorio siguiendo una distribución triangular utilizando el método de aceptación-rechazo.

    Args:
    a, b, c (float): Parámetros de la distribución triangular (a < b < c).

    Returns:
    float: Un número generado aleatoriamente siguiendo la distribución triangular.
    """
    while True:
        x = uniform(a, c)
        if x < b:
            acepta = (x - a) / (b - a)
        else:
            acepta = (c - x) / (c - b)
        if random() < acepta:
            break
    return x

# Generación y visualización de un histograma para la distribución triangular con el método de aceptación-rechazo
plt.hist([random_triangular(-3, 0, 8) for t in range(100000)], bins=200, density=True)
plt.show()
```
